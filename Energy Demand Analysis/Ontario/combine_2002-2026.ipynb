```python
# Source of data: https://reports-public.ieso.ca/public/Demand/
from pathlib import Path
import pandas as pd

yearly_folder = 'data/yearly_demand_2002-2026/'

csv_files = sorted(Path(yearly_folder).glob('PUB_Demand_*.csv'))
print(f"Found {len(csv_files)} CSV files:")
```

```python
dfs = []
total_rows = 0

for file in csv_files:
    df = pd.read_csv(file, skiprows=3) 
    dfs.append(df)
    rows = len(df)
    total_rows += rows
    print(f"Loaded {file.name:25} | Shape: {df.shape} | Rows: {rows}")

print(f"\n{'='*60}")
print(f"Total rows from all files: {total_rows}")
```

```python
combined_df = pd.concat(dfs, ignore_index=True)
print(combined_df.shape)
print(combined_df.head())
print(combined_df.dtypes)
print(combined_df.tail())
```

```python
# Save combined dataframe to CSV
output_file = './data/Combined_Hourly_Demand_2002-2026.csv'
combined_df.to_csv(output_file, index=False)
print(f"✓ Saved to: {output_file}")
print(f"Total rows saved: {len(combined_df)}")
```
